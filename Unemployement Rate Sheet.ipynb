```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
#Files

no_diploma = ("NO_HighSchool_Diploma-Copy1.csv")

high_only = ("HighSchool_Grad-Copy1.csv")

associate_deg = ("Associates_Degree-Copy1.csv")

bachelor_higher_deg = ("Bachelors_Or_Higher-copy1.csv")

#Import Data

dropout_dem = pd.read_csv(no_diploma)

high_dem = pd.read_csv(high_only)

associate_dem = pd.read_csv(associate_deg)

bachelor_dem = pd.read_csv(bachelor_higher_deg)
```

```python
# These Datasets showcase unemployement rate from 2009 to 2019 in different criteria of education; No Diploma, High School Diploma, Associates Degree, Bachelor's or Higher
# The Data also shows the age demographic of 25+
# participants that took the survery are in these 5 main industries; Business Services, Transportation, Medical, Manufacturing and Tech
# Survery concluded of 500,000 participants in major states
```

```python
plot_df = dropout_dem.melt(id_vars = "Year",var_name = "Month", value_name = "Unemployment %")
```

```python
figure(num=None, figsize=(16, 10), dpi=80, facecolor='w', edgecolor='k')
sns.lineplot(x = "Year", y = "Unemployment %", hue = "Month",data = plot_df)
```

```python
plot_df = high_dem.melt(id_vars = "Year",var_name = "Month", value_name = "Unemployment %")
```

```python
figure(num=None, figsize=(16, 10), dpi=80, facecolor='w', edgecolor='k')
p = sns.lineplot(x = "Year", y = "Unemployment %", hue = "Month",data = plot_df)
p.set_ylim(0, 16)
```

```python
figure(num=None, figsize=(16, 10), dpi=80, facecolor='w', edgecolor='k')
p = sns.lineplot(x = "Year", y = "Unemployment %", hue = "Month",data = plot_df)
p.set_ylim(0, 16)
```
